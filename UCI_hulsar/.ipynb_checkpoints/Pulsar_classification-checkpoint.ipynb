```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, KFold, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier
# from sklearn import metrics, feature_selection, preprocessing
```

```python
df0 = pd.read_csv('./data/HTRU2/HTRU_2.csv', 
                  names=('mean_int_pro','std_int_pro','kurt_int_pro','skew_int_pro',
                         'mean_dm_snr','std_dm_snr','kurt_dm_snr','skew_dm_snr','class'))
print(df0.head())
print(df0.isnull().sum())
print(df0.describe())
```

```python
#check the distributions of the continuous variables
fig1, axes = plt.subplots(nrows=1, ncols=8, sharey=False, figsize=(21, 2))
axes = axes.flat
cols = list(df0.columns)

for ax, name in zip(axes, cols[:-1]):
    ax.hist(df0[name], bins=12, edgecolor='white')
    ax.set_xlabel(name)
    print (name, ' skewness: ', scipy.stats.skew(df0[name]))

plt.show()
plt.close()
```

```python
#plot a pair grid
g = sns.PairGrid(data=df0, 
                 hue='class', 
                 hue_order=None, 
                 palette='husl', 
                 #hue_kws={"marker": ["o", 's', "D", '+']}, 
                 vars=cols[:-1],
                 x_vars=None, 
                 y_vars=None, 
                 diag_sharey=False, 
                 size=2.5, 
                 aspect=1, 
                 despine=True, 
                 dropna=True)
    
g = g.map_diag(plt.hist, edgecolor="white", linewidth=0.5)
g = g.map_offdiag(plt.scatter, edgecolor="white", linewidth=0.5, s=10)
g = g.add_legend()

sns.plt.show()
sns.plt.close()
```

```python
#grid search with KNN
params1 = {'n_neighbors':[3,6,9,12,15,18],
           'weights':['distance'],
           'algorithm':['auto']}
kf = KFold(n_splits=5, shuffle=True)

gs = GridSearchCV(estimator=KNeighborsClassifier(),param_grid=params1,cv=kf,scoring='accuracy')
mod1 = gs.fit(df0.drop('class', axis=1), df0['class'])

print(pd.DataFrame(mod1.cv_results_))
print(mod1.best_estimator_)
print('\n',mod1.best_score_)
```

```python
#grid search with logistic regression
params2 = {'solver':['liblinear'],
           'penalty':['l1','l2']}
kf = KFold(n_splits=5, shuffle=True)

gs = GridSearchCV(estimator=LogisticRegression(),param_grid=params2,cv=kf,scoring='accuracy')
mod2 = gs.fit(df0.drop('class', axis=1), df0['class'])

print(pd.DataFrame(mod2.cv_results_))
print(mod2.best_estimator_)
print('\n',mod2.best_score_)
```

```python
#grid search with SVC. SVC is very slow
params3 = {'kernel':['rbf','linear']}
kf = KFold(n_splits=5, shuffle=True)
gs = GridSearchCV(estimator=SVC(),param_grid=params3,cv=kf,scoring='accuracy')
mod3 = gs.fit(df0.drop('class', axis=1), df0['class'])
print(pd.DataFrame(mod3.cv_results_))
print(mod3.best_estimator_)
print('\n',mod3.best_score_)
```

```python
#grid search with LDA, very fast
params4 = {'solver':['svd','lsqr','eigen']}
kf = KFold(n_splits=5, shuffle=True)
gs = GridSearchCV(estimator=LinearDiscriminantAnalysis(),param_grid=params4,cv=kf,scoring='accuracy')
mod4 = gs.fit(df0.drop('class', axis=1), df0['class'])
print(pd.DataFrame(mod4.cv_results_))
print(mod4.best_estimator_)
print('\n',mod4.best_score_)
```

```python
#grid search with QDA, very fast.
params5 = {'reg_param':[0,0.2,0.4]}
kf = KFold(n_splits=5, shuffle=True)
gs = GridSearchCV(estimator=QuadraticDiscriminantAnalysis(),param_grid=params5,cv=kf,scoring='accuracy')
mod5 = gs.fit(df0.drop('class', axis=1), df0['class'])
print(pd.DataFrame(mod5.cv_results_))
print(mod5.best_estimator_)
print('\n',mod5.best_score_)
```

```python
#try sklearn.neural_network.MLPClassifier

kf = KFold(n_splits=5, shuffle=True)
param6 = {'activation':['logistic','relu'],
          'solver':['lbfgs'],
          'hidden_layer_sizes':(100,150,)}
gs = GridSearchCV(estimator=MLPClassifier(), 
                  param_grid=param6,
                  cv=kf,
                  scoring=None, 
                  return_train_score=True)
mod6 = gs.fit(df0.drop('class', axis=1), df0['class'])
print(pd.DataFrame(mod6.cv_results_))
print(mod6.best_estimator_)
print('\n','Best Score: ', mod6.best_score_)
```
